```python
import requests
import json
import random
from MySQLdb import _mysql
import time
import subprocess
import traceback

while True:
    try:
        db = _mysql.connect("localhost","root", "", "crawler_db")      
        for _ in range(100):
            try:
                flag = True
                print('before flag')
                while flag:
                    try:
                        db.query('CALL `PROCEDURE_URL_WIRTSCHAFT`(@p0, @p1); ')
                        db.query("SELECT @p0 AS `id`, @p1 AS `keyword`;")
                        flag = False
                    except Exception as e: 
                        print(e)
                        continue
                r=db.store_result()
                results=r.fetch_row()
                keyword = results[0][1].decode()
                print(keyword)

                time.sleep(random.uniform(1,3))
                url = f'https://www.wirtschaft.at/_next/data/2024.016.2/de/u/{keyword}.json?id={keyword}'
                response = requests.get(url)
                if response.status_code ==200:
                    with open(f"{keyword}.json", 'w', encoding='utf-8') as json_file:
                        json.dump(response.json(), json_file, indent=4)
                    print(f"{keyword}.json is saved")
                    for _ in range(5):
                        try:
                            query = f"UPDATE TEMP_URL_WIRTSCHAFT SET status = 10  WHERE company_id = '{keyword}'"
                            db.query(query)
                            break
                        except:
                            continue
                else:
                    print("No data or error")
                    pass
            except:
                pass
        
        try:
            db.close()
        except:
            pass

        try:
            subprocess.run(["aws", "s3", "mv", "./", "s3://dev_buk/DATA/SOURCE/WIRTSCHAFT/JSON/", "--recursive", "--exclude", "*", "--include", "*.json"], check=True)
        except: pass
    except:
        print(traceback.print_exc())
```

```python
from operator import itemgetter
import pandas as pd
import requests
import traceback

def process_keyword(keyword):
    print(keyword)
    url = f'https://www.wirtschaft.at/api/proxy/search/unternehmen?term={keyword}&page=1&limit=10000000'

    response = requests.get(url)
    response_json = response.json()
    # print(response_json)

    # Extract 'id' values using itemgetter and map
    ref_values = list(map(itemgetter('ref'), response_json['results']))
    ids = [ref['id'] for ref in ref_values]

    # Create a DataFrame
    df = pd.DataFrame({'Id': ids})

    # Specify the CSV file name
    csv_filename = 'company_ids.csv'

    # Append the DataFrame to the CSV file (mode='a' for append)
    df.to_csv(csv_filename, mode='a', header=False, index=False)

    print(f'Data for {keyword} has been successfully appended to {csv_filename}')

# Loop through lowercase letters ('a' to 'z')
for i in range(ord('a'), ord('z')+1):
    keyword = chr(i)
    process_keyword(keyword)

# Loop through digits (0 to 9)
for i in range(10):
    keyword = str(i)
    process_keyword(keyword)

# Loop through special characters
special_characters = ['!', '@', '#', '$', '%', '^', '&', '*','[',']','}','{', '.', ',', ';', ':', '/', '?', '>', '<','|', '(', ')', '-', '_', '+', '=', '~', '`']
for char in special_characters:
    keyword = char
    try:
        process_keyword(keyword)
    except:
        print(traceback.print_exc())

```

```python
import requests
import csv

keyword= 'aaaa'
url = f'https://www.wirtschaft.at/api/proxy/search/unternehmen?term={keyword}&page=1&limit=10000'  # 10000000

# Function to fetch data for a given 'i' value
def fetch_data(i):
    response = requests.get(url)
    response_json = response.json()
    cp_id = response_json['results'][i]['ref']['id']
    return cp_id

# Specify the CSV file name
csv_filename = f'{keyword}.csv'

# Open CSV file in write mode
with open(csv_filename, 'w', newline='') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)

    # Write header row
    csv_writer.writerow(['Company_ID'])  # Replace with actual column names

    # Loop through different values of 'i'
    i = 0
    while True:
        print(i)
        try:
            data = fetch_data(i)
            print(data)
        except:
            i += 1
            continue

        # Check if data is empty, exit the loop if there's no more data
        if not data:
            break

        # Write data to CSV
        csv_writer.writerow([data])  # Replace with actual field names

        # Increment 'i' for the next iteration
        i += 1

print(f'Data has been successfully written to {csv_filename}')
```
